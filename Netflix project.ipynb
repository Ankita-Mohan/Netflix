# Data Cleaning 

```python
#print the number of rows and columns
nf.shape
```

```python
#check for duplicate values
nf.duplicated().sum()
```

```python
#make a copy of the dataset
df=nf.copy()
```

```python
df.shape
```

```python
#check for NULL values
df.isnull().sum()
```

```python
#total NULL values
total_sum_null=df.isnull().sum().sum()
print("Total number of NULL values: ",total_sum_null)
```

```python
#check unique values
df.nunique()
```

```python
#drop NULL values
df=df.dropna()
df.shape
```

```python
#print starting 10 values
df.head(10)
```

```python
#convert date time format into date format for analysis
df["date_added"] = pd.to_datetime(df['date_added'], format='mixed')
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
#converting year and day into integer values
df['year_added'].astype(int);
df['day_added'].astype(int);
```

# Data Visualisation

```python
sns.countplot(x='type', data=nf) 
plt.title('Type Distribution')
plt.show()
```

```python
#rating of shows and movies
sns.countplot(x='rating', data=nf)  # Assuming 'nf' is your DataFrame
plt.title('Rating Distribution')
plt.show()
```

```python
#Relation between type and rating
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=nf)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
#pie chart for the type: Movie and TV shows
labels = ['Movie', 'TV show']
size = nf['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
#pie chart for rating
nf['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
